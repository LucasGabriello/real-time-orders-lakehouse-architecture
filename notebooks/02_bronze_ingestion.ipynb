```python
schema = spark.read.csv(
    "dbfs:/FileStore/orders-demo/source/orders.csv",
    header=True
).schema

bronze_df = (
    spark.readStream
    .format("cloudFiles")
    .option("cloudFiles.format", "csv")
    .option("header", "true")
    .schema(schema)
    .load("dbfs:/FileStore/orders-demo/landing")
)
```

```python
query = (
    bronze_df.writeStream
    .format("delta")
    .option("checkpointLocation", "/Volumes/workspace/default/orders_volume/checkpoints/bronze")
    .outputMode("append")
    .start("/Volumes/workspace/default/orders_volume/bronze")
)
```
