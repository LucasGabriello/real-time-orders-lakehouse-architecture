```python
import pandas as pd
import time

source = "/Volumes/workspace/default/orders_volume/source/orders.csv"
landing = "/Volumes/workspace/default/orders_volume/landing/"

df = pd.read_csv(source)

chunk_size = 10
num_files = 1

for file_num, i in enumerate(range(0, len(df), chunk_size)):
    
    if file_num >= num_files:
        break

    df[i:i+chunk_size].to_csv(
        f"{landing}orders_{file_num+1}.csv",
        index=False
    )

    print(f"Arquivo enviado: orders_{file_num+1}.csv")

    if file_num < num_files - 1:
        time.sleep(60)
```
