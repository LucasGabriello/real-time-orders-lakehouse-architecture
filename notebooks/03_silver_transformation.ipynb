```python
from pyspark.sql.functions import col

silver_df = (
    spark.readStream
    .format("delta")
    .load("/Volumes/workspace/default/orders_volume/bronze")
    .filter(col("Quantity") > 0)
    .filter(col("Price") > 0)
    .dropDuplicates(["TransactionNo"])
)
```

```python
silver_query = (
    silver_df.writeStream
    .format("delta")
    .option("checkpointLocation", "/Volumes/workspace/default/orders_volume/checkpoints/silver")
    .start("/Volumes/workspace/default/orders_volume/silver")
)
```
