```python
from pyspark.sql.functions import col, sum

gold_df = (
    silver_df
    .withColumn("revenue", col("Price") * col("Quantity"))
    .groupBy("Country")
    .agg(sum("revenue").alias("total_revenue"))
)
```

```python
gold_query = (
    gold_df.writeStream
    .format("delta")
    .outputMode("complete")
    .option("checkpointLocation", "/Volumes/workspace/default/orders_volume/checkpoints/gold")
    .start("/Volumes/workspace/default/orders_volume/gold")
)
```
